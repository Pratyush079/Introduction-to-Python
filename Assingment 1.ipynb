```python
# 2. Program that accepts temperature in Celsius and prints Fahrenheit and Kelvin values 
while True:
    try:
        c = float(input("What's the temperature in degree celsius? "))
        f = round(((c*9/5)+32),2)
        print(f"The temperature in Fahrenheit is {f}")
        k = round((c+273.15),2)
        print(f"The temperature in Kelvin is {k}")
        break
    except:
        print("Provide the temperature in celcius")
```

```python
# 3. Program that takes a number and checks whether it is positive, negative, integer, real or zero
try:
    n = complex(input("Give a number"))
    if n.imag == 0:
        print("Real Number")
        if n.real == 0:
            print("Zero")
        elif n.real > 0:
            print("Positive")
        else:
            print("Negative")
        if n.real.is_integer():
            print("Integer")
        else:
            print("Not an integer")
    else:
        print("Complex Number")
except:
    print("Not a number")
```

```python
# 4. Program to compute the sum of digits of a number. Skip the digit 5, and then find the factorial of the sum
while True:
    try:
        n = input("Give a number")
        s = 0
        f = 1
        if n.isdigit():
            for i in range(0,len(n)):
                if int(n[i]) == 5:
                    continue
                else:
                    s+= int(n[i])
            for i in range(1,s+1):
                f = f*i
            print(f"The sum of the digits is {s} and the factorial of the sum is {f}")
            break
    except:
        print("Provide a number")
```

```python
# 5. Write functions for addition, subtraction, multiplication, and division using a menu-driven approach, call the appropriate 
#    function based on user input
try:
    x = float(input("Give a number"))
    y = float(input("Give a number"))
except:
    print("Error! Not a number")


def a(x,y):
    return x+y
def s(x,y):
    return x-y
def m(x,y):
    return x*y
def d(x,y):
    if y == 0:
        return("Division by 0 is not possible")
    else:
        return x/y

try:
    print("Choose from options 1 to 4. Enter the option number")
    print("1.Addition \n2.Subtraction \n3.Multiplication \n4.Division")
    n = int(input())
    if n == 1:
        print(f"Result: {a(x, y)}")
    elif n == 2:
        print(f"Result: {s(x, y)}")
    elif n == 3:
        print(f"Result: {m(x, y)}")
    elif n == 4:
        print(f"Result: {d(x, y)}")
    else:
        print("Option not available")
except:
    print("Please enter a number")
```

```python
# 6. Write a function to reverse a number then sum the corresponding digits of both numbers and print them 

def rev(n,m):
    s = 0
    for i in range(len(n)):
        s = int(n[i]) + int(m[i])
        print("The sum of the corresponding digit is",s)

while True:
    try:
        n = input("Give a number")
        if n.isdigit():
            m = n[::-1]
            print(n,m)
            break
    except:
        print("Please input a number")
rev(n,m)
```

```python
# 7.Write a program that takes a string and prints how many times each character appears using a dictionary.
s = input("Enter a string:")
count = {}
for i in s:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
print("Character frequencies:")
for key,value in count.items():
    print(f"'{key}' appears {value} times")
```

```python
# 8.Given a list of integers, create a new list containing only the squares of the even numbers
try:
    l = []
    s = []
    n = int(input("Give the length of the list"))
    for i in range(n):
        x = int(input("Give a number: "))
        l.append(x)
        if x == 0:
            continue
        elif x%2 == 0:
            s.append(x**2)
    print("The previous list",l)
    print("The new list",s)
except:
    print("Please provide an integer as input")
```
