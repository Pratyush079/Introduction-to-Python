#### 2.	 Write a program to:
Generate the first 20 Fibonacci numbers, Store them in a list. Print only those greater than 50.

```python
a = 0
b = 1
l = []
for i in range(20):
    l.append(a)
    a, b = b, (a+b)
print("Numbers greater than 50 in the Fibonacci Series:")
for j in l:
    if j > 50:
        print(j, end = ' ')
```

#### 8. Generate synthetic data:
$y= 2x + 3 + noise $  
Fit a regression model  
Plot:
- data points 
- fitted line

```python
# x = np.linspace(1,100,50)
np.random.seed(38)
x = np.random.randint(1,100,50)
noise = np.random.normal(2,10,size = 50)
y = 2*x + 3 + noise

x = x.reshape(-1,1)

reg = LinearRegression()
reg.fit(x,y) # Fitting the model

y_pred = reg.predict(x) # Predicted values

# Plotting the points and fitting the regression line
plt.scatter(x,y, label = 'Data Points')
plt.plot(x,y_pred,color = 'red', label = 'Fitted Line')
plt.title('Regression of y on x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 9. Generate a sample of size 30 from a normal distribution  
Test: $H_0: μ = 50$  
Perform:
- one-sample t-test <br>
  Report: 
- test statistic
- p-value
- conclusion 

```python
np.random.seed(38)
mu = float(input('Give the mean of the normal distribution:'))
sigma = float(input('Give the standard deviation of the normal distribution:'))
print('The null hypothesis is μ = 50')
print('The alternate hypothesis is μ != 50')
n = np.random.normal(mu, sigma, size = 30)
t_stat, p_val = stats.ttest_1samp(n, 50)
print('\nThe test statistic:',round(t_stat,2),"\nThe p-value:",round(p_val,4))
if p_val < 0.05:
    print('We reject the null hypothesis')
else:
    print('We fail to reject the null hypothesis')
```

#### 10.	Create a time series data on yearly sales in an electronics store using Pandas. Perform a time series decomposition and try to determine the trend and seasonality.

```python
time = np.arange(96) # 8 years (96 months)

trend = 1.5*time
seasonal = 10 * np.cos(2 * np.pi * time / 8)
noise = np.random.normal(1, 5, len(time))

data = trend + seasonal + noise

df = pd.DataFrame({'Sales': data}, 
                  index=pd.date_range(start='2014-01-01', periods=96, freq='ME'))

result = seasonal_decompose(df['Sales'], model='additive', period=8)

result.plot()
plt.tight_layout()
plt.show()
```
